```python
# Algorithm to find the bollinger band
# total is the variable to sum the profits
# oneday is the variable for the number of 5mins in one day



ends = data['end']
times = data_date['date']
total = 0
count = 0
oneday = 71 

def bollinger_band(ends, oneday)
    for i in range(1, oneday):
        profit = ends[i] - ends[i-1]
        #print(times[i], ends[i], profit, profit**2)
        total += profit**2
    return total/oneday

print(total)
print(total/oneday)
    
```
